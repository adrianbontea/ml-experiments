```python
import os
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "D:\Projects\ml-experiments\datasets\housing\housing.csv"

data = pd.read_csv(csv_path)
```

Analyze the housing data

Quick peek at top 5 instances (districts):

```python
print(data.head())
```

Ocean Proximity counts in categories:

```python
 print(data.ocean_proximity.value_counts())
```

Description of the data set (percentiles etc):

```python
print(data.describe())
```

Slice a smaller DataFrame and plot:

```python
subset = data[["median_house_value", "median_income"]]
subset.hist()
```

Slice first 3 rows:

```python
subset = data[0:3]
print(subset)
```

Slice first 3 rows and 5 columns:

```python
subset = data.iloc[0:3, 0:5]
print(subset)
```

Query: Districts with Median Income == 10:

```python
subset = data[data.median_income == 10]
print(subset)
```

Plot the median housing value and population by latitude and longitude:

```python
data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
                            s=data.population / 100, label="population",
                            c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True)

plt.legend()
plt.show()
```

Find the correlations for median housing value based on correlation matrix:

```python
correlation_matrix = data.corr()
print(correlation_matrix["median_house_value"].sort_values(ascending=False))
```

The best correlation for predicting the median housing value seems to be with the median income attribute so plot their relation:

```python
data.plot(kind="scatter", x="median_income", y="median_house_value", alpha=0.1)
plt.show()
```
